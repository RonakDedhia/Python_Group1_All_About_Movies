```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt

df = pd.read_csv('imdb-5000-movies.csv')

#implementation for comparing genres by gross revenue

#create arrays for genre revenues - limited to top 10 genres 
comedy = []
adventure = []
action = []
drama = []
romance = []
thriller = []
western = []
horror = []
documentary = []
family = []

#create arrays for genre ratings - limited to top 10 genres 
comedy_rating = []
adventure_rating = []
action_rating = []
drama_rating = []
romance_rating = []
thriller_rating = []
western_rating = []
horror_rating = []
documentary_rating = []
family_rating = []

#create arrays for genre likes - limited to top 10 genres 
comedy_likes = []
adventure_likes = []
action_likes = []
drama_likes = []
romance_likes = []
thriller_likes = []
western_likes = []
horror_likes = []
documentary_likes = []
family_likes= []


#compiling all revenues, ratings, and facebook likes for each genre into an array        
for i in range(0,5043):
    if "Comedy" in df.genres[i]:
        if pd.notnull(df.gross[i]):
            comedy.append(df.gross[i])
        if pd.notnull(df.imdb_score[i]):
            comedy_rating.append(df.imdb_score[i])
        if pd.notnull(df.movie_facebook_likes[i]):
            comedy_likes.append(df.movie_facebook_likes[i])
    if "Adventure" in df.genres[i]:
        if pd.notnull(df.gross[i]):
            adventure.append(df.gross[i])
        if pd.notnull(df.imdb_score[i]):
            adventure_rating.append(df.imdb_score[i])
        if pd.notnull(df.movie_facebook_likes[i]):
            adventure_likes.append(df.movie_facebook_likes[i])
    if "Action" in df.genres[i]:
        if pd.notnull(df.gross[i]):
            action.append(df.gross[i])
        if pd.notnull(df.imdb_score[i]):
            action_rating.append(df.imdb_score[i])
        if pd.notnull(df.movie_facebook_likes[i]):
            action_likes.append(df.movie_facebook_likes[i])
    if "Drama" in df.genres[i]:
        if pd.notnull(df.gross[i]):
            drama.append(df.gross[i])
        if pd.notnull(df.imdb_score[i]):
            drama_rating.append(df.imdb_score[i])
        if pd.notnull(df.movie_facebook_likes[i]):
            drama_likes.append(df.movie_facebook_likes[i])
    if "Romance" in df.genres[i]:
        if pd.notnull(df.gross[i]):
            romance.append(df.gross[i])
        if pd.notnull(df.imdb_score[i]):
            romance_rating.append(df.imdb_score[i])
        if pd.notnull(df.movie_facebook_likes[i]):
            romance_likes.append(df.movie_facebook_likes[i])
    if "Thriller" in df.genres[i]:
        if pd.notnull(df.gross[i]):
            thriller.append(df.gross[i])
        if pd.notnull(df.imdb_score[i]):
            thriller_rating.append(df.imdb_score[i])
        if pd.notnull(df.movie_facebook_likes[i]):
            thriller_likes.append(df.movie_facebook_likes[i])
    if "Western" in df.genres[i]:
        if pd.notnull(df.gross[i]):
            western.append(df.gross[i])
        if pd.notnull(df.imdb_score[i]):
            western_rating.append(df.imdb_score[i])
        if pd.notnull(df.movie_facebook_likes[i]):
            western_likes.append(df.movie_facebook_likes[i])
    if "Horror" in df.genres[i]:
        if pd.notnull(df.gross[i]):
            horror.append(df.gross[i])
        if pd.notnull(df.imdb_score[i]):
            horror_rating.append(df.imdb_score[i])
        if pd.notnull(df.movie_facebook_likes[i]):
            horror_likes.append(df.movie_facebook_likes[i])
    if "Documentary" in df.genres[i]:
        if pd.notnull(df.gross[i]):
            documentary.append(df.gross[i])
        if pd.notnull(df.imdb_score[i]):
            documentary_rating.append(df.imdb_score[i])
        if pd.notnull(df.movie_facebook_likes[i]):
            documentary_likes.append(df.movie_facebook_likes[i])
    if "Family" in df.genres[i]:
        if pd.notnull(df.gross[i]):
            family.append(df.gross[i])
        if pd.notnull(df.imdb_score[i]):
            family_rating.append(df.imdb_score[i])
        if pd.notnull(df.movie_facebook_likes[i]):
            family_likes.append(df.movie_facebook_likes[i])      



#computing average revenue for each genre
comedy_mean = np.mean(comedy)
adventure_mean = np.mean(adventure)
action_mean = np.mean(action)
drama_mean = np.mean(drama)
western_mean = np.mean(western)
romance_mean = np.mean(romance)
thriller_mean = np.mean(thriller)
horror_mean = np.mean(horror)
documentary_mean = np.mean(documentary)
family_mean = np.mean(family)

#create bar graphs showing differences in average revenue between 10 genres 

 
objects = ('Comedy', 'Adventure', 'Action', 'Drama', 'Western', 'Romance', 'Thriller', 'Horror', 'Documentary', 'Family')
y_pos = np.arange(len(objects))
revenue = [comedy_mean,adventure_mean,action_mean,drama_mean,western_mean,romance_mean,thriller_mean,horror_mean,documentary_mean,family_mean]
 
plt.barh(y_pos, revenue, align='center', alpha=.5)
plt.yticks(y_pos, objects)
plt.xlabel('Revenue in Hundred Millions')
plt.title('Movie Revenue by Genre')
 
plt.show()

#computing average rating for each genre
comedy_rating_mean = np.mean(comedy_rating)
adventure_rating_mean = np.mean(adventure_rating)
action_rating_mean = np.mean(action_rating)
drama_rating_mean = np.mean(drama_rating)
western_rating_mean = np.mean(western_rating)
romance_rating_mean = np.mean(romance_rating)
thriller_rating_mean = np.mean(thriller_rating)
horror_rating_mean = np.mean(horror_rating)
documentary_rating_mean = np.mean(documentary_rating)
family_rating_mean = np.mean(family_rating)

#create bar graphs showing differences in average IMDB rating between 10 genres 

 
objects = ('Comedy', 'Adventure', 'Action', 'Drama', 'Western', 'Romance', 'Thriller', 'Horror', 'Documentary', 'Family')
y_pos = np.arange(len(objects))
rating = [comedy_rating_mean,adventure_rating_mean,action_rating_mean,drama_rating_mean,western_rating_mean,romance_rating_mean,thriller_rating_mean,horror_rating_mean,documentary_rating_mean,family_rating_mean]
 
plt.barh(y_pos, rating, align='center', alpha=.5)
plt.yticks(y_pos, objects)
plt.xlabel('Rating')
plt.title('Movie IMDB Rating by Genre')
 
plt.show()

#computing average likes for each genre
comedy_likes_mean = np.mean(comedy_likes)
adventure_likes_mean = np.mean(adventure_likes)
action_likes_mean = np.mean(action_likes)
drama_likes_mean = np.mean(drama_likes)
western_likes_mean = np.mean(western_likes)
romance_likes_mean = np.mean(romance_likes)
thriller_likes_mean = np.mean(thriller_likes)
horror_likes_mean = np.mean(horror_likes)
documentary_likes_mean = np.mean(documentary_likes)
family_likes_mean = np.mean(family_likes)

#create bar graphs showing differences in average Facebook likes between 10 genres 

 
objects = ('Comedy', 'Adventure', 'Action', 'Drama', 'Western', 'Romance', 'Thriller', 'Horror', 'Documentary', 'Family')
y_pos = np.arange(len(objects))
likes = [comedy_likes_mean,adventure_likes_mean,action_likes_mean,drama_likes_mean,western_likes_mean,romance_likes_mean,thriller_likes_mean,horror_likes_mean,documentary_likes_mean,family_likes_mean]
 
plt.barh(y_pos, likes, align='center', alpha=.5)
plt.yticks(y_pos, objects)
plt.xlabel('Likes')
plt.title('Movie Facebook Likes by Genre')
 
plt.show()
```
