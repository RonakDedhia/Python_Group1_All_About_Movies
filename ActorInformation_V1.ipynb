```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('imdb-5000-movies.csv')

actor = str(input('What Actor or Actress Would You Like Information On?: '))

#check that actor is in database & count how many movies they are in
count = 0
count_1 = 0
count_2 = 0
count_t = 0
for i in range(0,5043):
    if actor == df.actor_1_name[i]:
        count = count + 1
        count_t = count_t + 1
    if actor == df.actor_2_name[i]:
        count = count + 1
        count_1 = count_1 + 1
    if actor == df.actor_3_name[i]:
        count = count + 1
        count_2 = count_2 + 1
        
#output if they are not in the database
if count == 0:
    print('Sorry this actor/actress is not in the database!')
else:
    print(str(actor)+' is in '+str(count)+' movies in this database.')
    
#compile movies that the actor is actor 1, 2 and 3 in & add their imdb scores & their genres
actor_1_movies = []
actor_2_movies = []
actor_3_movies = []
ratings = []
actor_1_genres = []
actor_2_genres = []
actor_3_genres = []
actor_1_count = 0
for i in range(0,5043):
    if actor == df.actor_1_name[i]:
        actor_1_movies.append(df.movie_title[i])
        actor_1_count = actor_1_count + 1
        facebook = df.actor_1_facebook_likes[i]
        ratings.append(df.imdb_score[i])
        if pd.notnull(df.genres[i]):
            actor_1_genres.append(df.genres[i])        
    if actor == df.actor_2_name[i]:
        actor_2_movies.append(df.movie_title[i])
        ratings.append(df.imdb_score[i])
        if pd.notnull(df.genres[i]):
            actor_2_genres.append(df.genres[i])
    if actor == df.actor_3_name[i]:
        actor_3_movies.append(df.movie_title[i])
        ratings.append(df.imdb_score[i])
        if pd.notnull(df.genres[i]):
            actor_3_genres.append(df.genres[i])
            
#Outputting information on the actor/actress inputted
print("\n"+str(actor)+' is the main actor/actress in the following movies:')
print(*actor_1_movies, sep = "\n") 

if count_1 != 0:
    print("\n"+str(actor)+' is the second main actor/actress in the following movies:')
    print(*actor_2_movies, sep = "\n")  

if count_2 != 0:
    print("\n"+str(actor)+' is the third actor/actress in the following movies:')
    print(*actor_3_movies, sep = "\n")   
    
print("\n"+str(actor+' has '+str(facebook)+' facebook likes.'))

rating_mean = np.mean(ratings)

print("\n"+str(actor)+' has an average imdb rating of '+str(rating_mean))

#count number of movies from each genre that the actot/actress is in
comedy = 0
adventure = 0
action = 0
drama = 0
romance = 0
thriller = 0
western = 0
horror = 0
documentary = 0
family = 0
        
for i in range(0,count_t):
    if 'Comedy' in actor_1_genres[i]:
        comedy = comedy + 1
    if 'Adventure' in actor_1_genres[i]:
        adventure = adventure + 1
    if 'Action' in actor_1_genres[i]:
        action = action + 1
    if 'Drama' in actor_1_genres[i]:
        drama = drama + 1
    if 'Romance' in actor_1_genres[i]:
        romance = romance + 1
    if 'Thriller' in actor_1_genres[i]:
         thriller = thriller + 1
    if 'Western' in actor_1_genres[i]:
        western = western + 1
    if 'Horror' in actor_1_genres[i]:
        horror = horror + 1    
    if 'Documentary' in actor_1_genres[i]:
        documentary = documentary + 1
    if 'Family' in actor_1_genres[i]:
        family = family + 1   
    if count_1 != 0:
        for i in range(0,count_1):
            if 'Comedy' in actor_2_genres[i]:
                comedy = comedy + 1
            if 'Adventure' in actor_2_genres[i]:
                adventure = adventure + 1
            if 'Action' in actor_2_genres[i]:
                action = action + 1
            if 'Drama' in actor_2_genres[i]:
                drama = drama + 1
            if 'Romance' in actor_2_genres[i]:
                romance = romance + 1
            if 'Thriller' in actor_2_genres[i]:
                thriller = thriller + 1
            if 'Western' in actor_2_genres[i]:
                western = western + 1
            if 'Horror' in actor_2_genres[i]:
                horror = horror + 1
            if 'Documentary' in actor_2_genres[i]:
                documentary = documentary + 1
            if 'Family' in actor_2_genres[i]:
                family = family + 1
    if count_2 != 0:
        for i in range(0,count_2):
            if 'Comedy' in actor_3_genres[i]:
                comedy = comedy + 1
            if 'Adventure' in actor_3_genres[i]:
                adventure = adventure + 1
            if 'Action' in actor_3_genres[i]:
                action = action + 1
            if 'Drama' in actor_3_genres[i]:
                drama = drama + 1
            if 'Romance' in actor_3_genres[i]:
                romance = romance + 1
            if 'Thriller' in actor_3_genres[i]:
                thriller = thriller + 1 
            if 'Western' in actor_3_genres[i]:
                western = western + 1
            if 'Horror' in actor_3_genres[i]:
                horror = horror + 1  
            if 'Documentary' in actor_3_genres[i]:
                documentary = documentary + 1
            if 'Family' in actor_3_genres[i]:
                family = family + 1

#create pie chart that shows genre distribution

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Comedy', 'Adventure', 'Action', 'Drama','Romance','Thriller','Western','Horror','Documentary','Family'
sizes = [family, adventure, action, drama, romance, thriller, western, horror, documentary, family]



fig1, ax1 = plt.subplots()
text = ax1.pie(sizes, labels=labels, autopct='%1.f%%', startangle=0)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.


plt.show()        
```
